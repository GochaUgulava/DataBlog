### 2. მონაცემთა ტიპები

<p style='text-align: justify;'>როგორც ვიცით, მონაცემები შეიძლება მრავალგვარი იყოს. თუმცა, შესაძლებელია გარკვეული ნიშან-თვისებებით მათი დაჯგუფება. უპირველეს ყოვლისა გავმიჯნოთ რაოდენობრივი და თვისებრივი მონაცემები. </p><br>
<p style='text-align: justify;'><b>რაოდენობრივი მონაცემები</b> რაიმეს გაზომვით ან დათვლით მიიღება. თავის მხრივ თუ მონაცემი მოცემულ მონაკვეთზე   სკალის გამოყენებით გაზომვის შედეგია ის უწყვეტია. უწყვეტი მონაცემები შეიძლება წილადების სახითაც წარმოვიდგინოთ, ისინი მდორედ და წყვეტის გარეშე იცვლებიან. ხოლო რაოდენობრივი <b>დისკრეტული მონაცემები</b> ღებულობს მთელ რიცხვით მნიშვნელობებს, მათ წილადი ნაწილი არ გააჩნიათ. მაგალითად რაოდენობრივი უწყვეტი მონაცემებია: ტემპერატურის ცვლილების მაჩვენებლები ცელსიუსის გრადუსებში, ქვეყნების მიხედვით მთლიანი შიდა პროდუქტის მოცულობები ფულად ერთეულში გამოსახული, სკოლაში მოსწავლეთა სიმაღლეების მნიშვნელობები მეტრებში. რაოდენობრივი დისკრეტული მნიშვნელობებია: სტუდენტთა რაოდენობა ჯგუფში, მაღაზიაში მყიდველთა რაოდენობა, ცვლის განმავლობაში ჩარხზე დამზადებული დეტალების რაოდენობა.  მათემატიკურ ანალოგიას თუ გავაკეთებთ, ყველაზე მეტად უწყვეტ მონაცემებს ნამდვილი რიცხვები შეესაბამება, ხოლო დისკრეტულ მონაცემებს კი ნატურალური რიცხვები. უწყვეტ მონაცემებში გამოვყოთ ინტერვალური და ფარდობითი მონაცემები. <b>ინტერვალური მონაცემებია</b> - გაზომვადი, ფიზიკური აზრის მქონე, აბსოლუტური მონაცემები. ხოლო, <b>ფარდობითი მონაცემებია</b> -  განზომილების არ მქონე, წილობითი ან პროცენტული მონაცემები.</p><br>
<p style='text-align: justify;'><b>თვისებრივი მონაცემები</b> რაიმე თვისებას, თვისებრივ მახასიათებლებს ან მდგომარეობას გამოხატავენ. თვისებრივ მონაცემებში განვასხვავებთ ნომინალურ და რიგობრივ მონაცემებს. <b>ნომინალური მონაცემები</b> ობიექტის თვისებებს რაიმე ნიშნით  გამოკვეთს ან გამოსახავს მისი რომელიმე კატეგორიის ცვლილებას. <b>რიგობრივი მონაცემები</b> კი გვაჩვენებს რაიმე ნიშნის  ხარისხობრივი გამოხატულების დონეს. რიგობრივი მონაცემები რანჟირებულია, შესაძლებელია მათი რაიმე ნიშნით დალაგება. ნომინალური მაჩვენებლების მოწესრიგება, მათი  დალაგება შეუძლებელია. ნომინალური მაჩვენებლებია: სათამაშოების ფერები (წითელი, ყვითელი, მწვანე, ლურჯი, თეთრი ...), ავტომობილები გადაცემათა კოლოფის მიხედვით (ავტომატური, მექანიკური). რიგობრივი მაჩვენებლების მაგალითებია: სამხედრო წოდებები (ლეიტენანტი, კაპიტანი, მაიორი ...), მაისურის ზომები (S, M, L, XL, XXL …), შესწავლილი მასალის გაგების დონე (ძალიან ცუდი, ცუდი, საშუალო, კარგი, ძალიან კარგი). რიგობრივ მაჩვენებლებს შესაძლებელია რიცხობრივი შესატყვისები შევუფარდოთ, მაგალითად -2 ძალიან ცუდი, -1 ცუდი, 0 საშუალო, 1 კარგი, 2 ძალიან კარგი. ნომინალური მონაცემები რიცხვებისაგან კიდევ უფრო შორს დგანან. მათთვის რაიმე რიცხვების შესატყვისება სულაც არ გულისხმობს მათ მოწესრიეგებას, რანჟირებას. მაგალითად ავტომატური გადაცემათა კოლოფისათვის კოდი 1-ის მინიჭება, ხოლო მექანიკური გადაცემათა კოლოფისათვის კოდი 2-ის მინიჭება სულაც არ გულისხმობს ერთის მეტობას ან უკეთესობას მეორეზე. მიუხედავათ რიცხვობრივი გამოსახვის შესაძლებლობისა ჩვენ არ შეგვიძლია თვისობრივ მონაცემებს შორის „მანძილების“, ინტერვალების გაზომვა - მათთვის არ არსებობს ინტერვალური სკალები.</p><br>
<p style='text-align: justify;'>ცალკე აღნიშვნას იმსახურებს <b>ბინარული მონაცემები</b>. ისინი მხოლოდ ორ მნიშვნელობას იღებენ. მაგალითად: 0 ან 1,  კი ან არა, TRUE ან FALSE და ა. შ. შესაძლებელია კიდევ სხვა ტიპის მონაცემებიც შეგვხვხდეს, რომლებიც ასახავენ დროს, თარიღებს, კუთხეებს და სხვა სპეციფიურ ელემენტებს. თუმცა მონაცემთა ძირითადი მოცულობა მოცემულ კლასიფიკაციაში ჯდება.</p><br>
<p style='text-align: justify;'>პრაქტიკაში მონაცემთა რიგი ტიპების განსაზღვრისას შეიძლება გადავეყაროთ გარკვეულ სირთულეებს. ხშირად ძნელი გასამიჯნია ერთმანეთისაგან რაოდენობრივი დისკრეტული და თვისებრივი რიგობრივი მონაცემები. ამ შემთხვევაში უნდა დავაკვირდეთ თითოეულის  გაზომვის სკალას. თუ სკალაზე მნიშვნელობებს შორის მანძილები მკვეთრად გამოხატულად ტოლია, მაშინ რაოდენობრივ დისკრეტულ მონაცემებთან გვაქვს საქმე. თვისებრივ რიგობრივი მონაცემებისათვის სკალის მნიშვნელობებს შორის სხვადასხვა მანძილებია. ასევე, ზოგჯერ რთულია რაოდენობრივ უწყვეტ და დისკრეტულ მაჩვენებლებს შორის ზღვარის გავლება. მიღებულია,  რომ თუ რაოდენობრივი მაჩვენებელი დისკრეტული ნიშნის გრადაციაში 20-ზე მეტ მნიშვნელობას შეიცავს ის შეიძლება რაოდენობრივ უწყვეტ მაჩვენებლად ჩავთვალოთ.</p><br>
<p style='text-align: justify;'>ხშირად პრაქტიკაში შესაძლებელია სხვადასხვა საჭიროებებიდან გამომდინარე მონაცემთა ტიპის შეცვლა. მაგალითად თუ რაიმე ობიექტის ზომის აღწერისას ნაცვლად თვისებრივი რიგობრივი მონაცემებისა - პატარა, საშუალო, დიდი - მოვახდენთ გაზომვას და ჩავინიშნავთ მონაცემებს ზომების კონკრეტული ციფრობრივი მითითებით მივიღებთ რაოდენობრივ ინტერვალურ მონაცემებს.  ანალოგიურად, თვისებრივ ნომინალურ სიდიდეს (მაგლითად ფერებს) თუ ავღრიცხავთ RGB სისტემის მიხედვით და შევუსაბამებთ ციფრობრივ მნიშვნელობებს მივიღებთ რაოდენობრივ დისკრეტულ მონაცემებს, ან ფერის ტალღის სიგრძის სპექტროფოტომეტრით გაზომვის შემთხვევაში - რაოდენობრივ უწყვეტ მონაცემებს. ასევე იოლია რაოდენობრივი მონაცემების თვისებრივ რიგობრივ მონაცემებათ გარდაქმნა. ამისათვის საკმარისია მოვახდინოთ მათი გრადაცია და დავაწესოთ შეაბამისი საზღვრები, შემდეგ შევუსაბამოთ თავ-თავისი რიგობრივი მაჩვენებლები. მაგალითად ადამიანთა სიმაღლე რაოდენობრივი უწყვეტი მონაცემებიდან გადავიყვანოთ თვისებრივ რიგობრივ მონაცემებში: შევუსაბამოთ 160 სმ-მდე სიმაღლე - დაბალს, 160 სმ-დან 180 სმ-მე - საშუალოს და 180-ს ზემოდ სიმაღლე - მაღალს, შედეგად მონაცემები მიიღებს თვისებრივ რიგობრივ სახეს.</p><br>
