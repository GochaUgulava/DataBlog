### 1. რა არის მონაცემები? გენერალური ერთობლიობა და შერჩევა

<p style='text-align: justify;'>მონაცემების საყოველთაოდ აღიარებულ, ერთმნიშვნელოვან განმარტებას ვერ ვიპოვით. მეცნიერების ყოველი დარგი თავისი სპეციფიკით უდგება და ახასიათებს მონაცემებს, ყოფით პრაქტიკაში კი ხშირად ვსარგებლობთ ამ ცნებით, მისი ინტუიციურ დონეზე ჩამოყალიბებული შინაარსით,   ყოველგვარი  განმარტების გარეშე. ზოგადი სახით, მონაცემები - ინფორმაციაა (მოვლენები, ფაქტები, შეტყობინებები, მახასიათებლები, სიგნალები), რომელიც გამოიყენება ადამიანის ან ტექნიკური სისტემის მიერ მათთვის მისაღები ფორმით დამუშავებისათვის, წარმოდგენისათვის, ანალიზისათვის, ინტერპრეტაციისათვის, აღქმისათვის, შენახვისა და გადაცემისათვის. ინფორმაცია - ობიექტურია, ის დამკვირვებლისაგან დამოუკიდებლად არსებობს, თუმცა მონაცემებად მაშინ იქცევა, როდესაც დამკვირვებელი მოახდენს მათ მიზანმიმართულ შეგროვებას. მიუხედავათ ფორმისა, ინფორმაცია უპირველეს ყოვლისა შინაარსობრივი არსის მატარებელია. მონაცემები კი გარკვეულ ფორმაში მოქცეული ინფორმაციაა.</p> <br>
<p style='text-align: justify;'>ინტერნეტის განვითარებამ მონაცემთა გენერირების არნახული ზრდა გამოიწვია. 1993 წელს 130 საიტი არსებობდა, დღეისათვის მათი რიცხვი 1,2 მილიარდს აჭარბებს.  Google-ის მიერ ერთ დღეში  დამუშავებული  ინფორმაციის მოცულობა მეტია ვიდრე ინტერნეტის ეპოქამდე მსოფლიოს ყველა ენაზე დაწერილი ყველა ლიტერატურული ნაშრომის მოცულობა. Youtube-ის მონაცემების მოცულობა 5 პეტაბაიტია (პეტა - 10<sup>15</sup> ). ყოველ წუთს ფეისბუქის მომხმარებლები ქსელში ტვირთავენ  500 000 კომენტარს, 300 000 სტატუსს, 130 000 ფოტოს. ამ მონაცემების შესანახად კომპანიები აშენებენ დატა-ცენტრებს, თითოეულს საშუალოდ  რამოდენიმე  ათასი კვადრატული მეტრის ფართობი უჭირავს. როგორც Google-ის დირექტორთა საბჭოს თავმჯდომარემ ერიკ შმიდტმა განაცხადა: „ცივილიზაციის გაჩენის მომენტიდან 2003 წლამდე პერიოდში კაცობრიობამ შექმნა 5 ექსაბაიტის (ექსა - 10<sup>18</sup>) მონაცემები. ეხლა უკვე,  ყოველ ორ დღეში ვქმნით 5 ექსაბაიტ მონაცემს, და ეს სიჩქარე კიდევ უფრო იზრდება“.  ჩვენს ყოველდღიურობაში ტერმინმა „დიდმა მონაცემებმა“ – “Big Data” მტკიცედ დაიმკვიდრა ადგილი. „დიდი მონაცემები“  XXI საუკუნის უმთავრეს რესურსად იქცა, მას ჩვენი დროების „ნავთობსაც“ კი უწოდებენ, უდავოა რომ ისინი  ცვლიან ადამიანების ცხოვრებას.</p><br>
<p style='text-align: justify;'>ანალიზისათვის მონაცემთა მიღების მრავალი გზა არსებობს. მათ შორის ძირითადებია დაკვირვება და  ექსპერიმენტი. განვასხვავოთ გენერალური ერთობლიობის და შერჩევის ცნებები. <b>გენერალური ერთობლიობა</b> - ყველა იმ ობიექტის ერთობლიობაა, რომელთა შესწავლაც ხორციელდება, რომლებზეც უნდა ჩატარდეს დაკვირვება. ხშირად გენერალური ერთობლიობის ობიექტთა რაოდენობა იმდენად დიდია (ზოგჯერ უსასრულოც), რომ არარეალურია მის უკლებლივ ყველა ობიექტზე დაკვირვების წარმოება. ამ შემთხვევაში ვაკვირდებით (ან ექსპერიმენტს ვუტარებთ) გენერალური ერთობლიობის  მხოლოდ ნაწილ ობიექტებს. გენერალური ერთობლიობიდან დაკვირვებისათვის (ექსპერიმენტისათვის) სპეციალური პროცედურების დაცვით შერჩეულ ობიექტთა ერთობლიობას ეწოდება <b>შერჩევითი ერთობლიობა, ან შერჩევა</b>. შერჩევის ელემენტთა ანლიზის შედეგად მიღებული დასკვნები შესაძლებელია განვავრცოთ მთლიანად გენერალურ ერთობლიობაზე. ოღონდ, აქვე უნდა გავაკეთოთ ერთი დათქმა: შერჩევა აუცილებალად უნდა იყოს რეპრეზენტატიული ანუ საკმარისად წარმომადგენლობითი. შერჩევის რეპრეზენტატიულობის უზრუნველსაყოფად შერჩევისათვის მონაცემთა აკრების დროს აუცილებლად უნდა დავიცვათ ორი პრინციპი: განმეორადობა და რანდომიზაცია.</p><br> 
<p style='text-align: justify;'>განმეორადობა გულისხმობს, რომ დაკვირვება გენერალური ერთობლიობის რამდენიმე ობიექტზე უნდა განმეორდეს, მხოლოდ ერთ ობიექტზე დაკვირვება შედეგს ვერ მოგვცემს. განმეორებადობები ერთმანეთისაგან დამოუკიდებელნი უნდა იყვნენ. მრავალი მოსაზრება არსებობს თუ რამდენჯერ უნდა განმეორდეს დაკვირვება. რა თქმა უნდა რაც მეტჯერ - მით უკეთესი, თუმცა არსებობს სპეციალური ტესტები (სიმძლავრის ტესტები), რომლებიც ადგენს თუ რამდენი მონაცემია საჭირო მოცემული სიზუსტით რაიმე დასკვნის გასაკეთებლად. პრაქტიკაში, ამ სირთულეების თავიდან ასაცილებლად არსებობს ემპირიულად დადგენილი „30-ის წესი“. 30 ელემენტზე ნაკლები შემცველობის  შერჩევებს - მცირე შერჩევებს უწოდებენ, 30 ელემენტზე მეტ შემცველ შერჩევებს - დიდ შერჩევებს. თითოეულ ტიპთან სამუშაოდ თავ-თავისი პროცედურებია შემუშავებული.</p> <br>
<p style='text-align: justify;'>აუცილებელია შერჩევაში თითოეული ელემენტი შემთხვევითი შერჩევის წესით აღმოჩნდეს, რასაც მეორე პრინციპი - რანდომიზაცია მოითხოვს. გენერალური ერთობლიობის თითოეულ ელემენტს აბსოლუტურად ერთი და იგივე შანსი უნდა გააჩნდეს რაც მის სხვა ელემენტებს შერჩევაში მოსახვედრად. პრაქტიკაში ძალზე დიდი ძალისხმევა და ყურადღებაა საჭირო ამ ორი პრინციპის განუხრელად დასაცავად. რეპრეზენტატიულობის დარღვევამ შეიძლება არასწორი დასკვნები გაგვაკეთებინოს. ამის მაგალითად მოვიყვანთ აშშ-ში 1936 წლის საპრეზიდენტო არჩევნებისათვის ჟურნალ „The Literary Digest“-ის მიერ ჩატარებულ  სოციოლოგიურ კვლევას.  ჟურნალმა წინასაარჩევნოდ გამოკითხა თავისი ხელმომწერები თუ რომელ საპრეზიდენტო კანდიდატს უჭერდნენ მხარს. რადგანაც ჟურნალის მესვეურებმა იცოდნენ, რომ მათი ხელმომწერების უმეტესობა რესპუბლიკელების მხარდამჭერი იყო, რეპრეზენტატულობის გასაზრდელად შემთხვევითი შერჩევის მეთოდით სატელეფონო ცნობარიდან და ავტომობილების მფლობელთა სარეგისტრაციო სიებიდან დაამატეს საკმაო რაოდენობის გამოკითხული. გამოკითხვის შედეგებით, მომავალ არჩევნებში რესპუბლიკელთა კანდიდატს ლენდონს 60% მხარდამჭერით მოქმედი პრეზიდენტის რუზველტის 40%-ის პირისპირ გამარჯვების ყველანაირი შანსი გააჩნდა. თუმცა, როგორც ვიცით ყველაფერი ზუსტად საპირისპიროდ წარიმართა. შეცდომა გამოიწვია შერჩევის არარეპრეზენტატულობამ. ჟურნალის მესვეურებმა ვერ გაითვალისწინეს არსებული რეალიები, კერძოდ ის, რომ დიდი დეპრესიის წლებში ტელეფონისა და ავტომობილის ფლობის შესაძლებლობა მდიდარ, შეძლებულ ხალხს გააჩნდა, რომელთა აბსოლუტური უმრავლესობა რესპუბლიკელებს უჭერდა მხარს. შედეგად დაირღვა შერჩევის ჩამოყალიბების რადომიზაციის პრინციპი, რამაც დასკვნების მცდარობა გამოიწვია.</p><br>
