### მოგესალმებით!

<p style='text-align: justify;'>ამ ბლოგში შევეცდებით მოგაწოდოთ ინფორმაცია ისეთი საინტერესო თემის შესახებ, როგორიც მონაცემთა ანალიზია. დღეს ძალიან პოპულარული გახდა და ხშირადაც წააწყდებით ინტერნეტში, სოციალურ ქსელებში თუ ტელევიზიაში ცნებებს „მონაცემები“, „მონაცემთა ანალიზი“, „დიდი მონაცემები“, „მონაცემთა ინჟინერი“, „მონაცემთა მეცნიერი“, „მანქანური სწავლება“, „ღრმა სწავლება“, „ნეირონული ქსელები“, „ხელოვნური ინტელექტი“  და მრავალი სხვა. ამ ბლოგის მეშვეობით ჩვენ გადავდგამთ პირველ ნაბიჯებს და შევაღებთ კარებს საინტერესო სამყაროში, რომელსაც მონაცემთა ანალიზი ეწოდება. ჩვენი მიზანია მოგაწოდოთ პირველადი, საწყისი ინფორმაცია მონაცემების, მათი დამუშავების და ანალიზის თაობაზე. ბლოგი არ იქნება გადატვირთული მათემატიკური ფორმულებით, მტკიცებებით, თეორემებით. ინფორმაციას წმინდა პრაქტიკული დანიშნულება ექნება. მისი გამოყენებით ადამიანმა შეძლოს ჩაატაროს მონაცემთა უმარტივესი ანალიზი: დაამუშავოს საწყისი მონაცემები, გაანალიზოს სტატისტიკური ინსტრუმენტების (ტესტების) მეშვეობით, მიიღოს შედეგები, მოახდინოს მათი ინტერპრეტირება და მარტივი ვიზუალიზაცია. ამ მასალის შესწავლის შემდგომ, თემის სრულ დაუფლებამდე კიდევ ძალზე დიდი გზა იქნება გასავლელი, თუმცა პირველი ნაბიჯები უკვე გადადგმული გექნებათ, თავს უცხოდ არ იგრძნობთ და შემდგომ სვლასაც, დაინტერესების შემთხვევაში, დამოუკიდებლად შეძლებთ. სტატისტიკური მეცნიერება მრავალ ინსტრუმენტს გვთავაზობს მონაცემთა დასამუშავებლად, ჩვენ განვიხილავთ მათ მცირე, თუმცა ეფექტურ ნაწილს, რათა ამ არსენალით აღჭურვილებს შეგეძლოდ ჩაატაროთ მონაცემთა კვლევა და ანალიზი, მიიღოთ დასკვნები. ჩვენ არ შევეხებით ამ ტესტების მათემატიკურ დასაბუთებას, მათი გამოყვანის მკაცრ მტკიცებებს. (ამ საკითხით დაინტერესებულ პირებს ვურჩევთ ალბათობის თეორიის, მათემატიკური სტატისტიკის, ეკონომეტრიკის, პროგნოსტიკის მრავლად არსებულ ლიტერატურას). ბლოგის მასალა მხოლოდ პრაქტიკული, გამოყენებითი ხასიათის იქნება. ჩვენ გავამახვილებთ ყურადღებას სტატისტიკური ტესტების პრაქტიკაში გამოყენების პირობებზე, მაგალითებზე, მიღებული შედეგების ინტერპრეტირებაზე.</p><br>
<p style='text-align: justify;'>პირველი საწყისი თემებით გავეცნობით მონაცემთა ანალიზის უზოგადეს ცნებებს, პრინციპებს. ეს ინფორმაცია მთელი კურსის განმავლობაში ჰაერივით საჭირო იქნება. ამიტომ, ცნებებში გარკვევა, მათი გააზრებული გამოყენება  ძალზე მნიშვნელოვანია. ანალიზის წარმოება მრავალი მეთოდითაა შესაძლებელი.   ტესტები ქაღალდისა და ფანქრის მეშვეობით შეიძლება იქნას ჩატარებული, თუმცა დღეს ეს უკვე წარმოუდგენელიც კი არის. ინსტრუმენტული საშუალებების, გამოთვლითი ტექნიკის, სპეციალიზებული პროგრამების განვითარებამ სრულიად ახალი დატვირთვა შეძინა მონაცემთა ანალიზს. მოსაწყენი, რუტინული გაანგარიშებები კომპიუტერმა იტვირთა თავის თავზე. ანალიტიკოსს გამოუნთავისუფლა მეტი დრო შემოქმედებითი საქმიანობისათვის - მონაცემთა ურთიერთკავშირებისა და სტრუტქტურის საძიებლად, მოდელების ასაგებად, კანონზომიერებების დასადგენად, შედეგების ინტერპრეტაციისათვის. პროგრამულმა პაკეტებმა იტვირთა ასევე გასაოცარი ვიზუალიზაციების აგებაც. წარმოუდგენლად მრავალი მეთოდი გაჩნდა მონაცემების, შედეგების გრაფიკულად გამოხატვისათვის, რამაც შეუდარებლად გაუადვილა  მოუმზადებელ ადამიანებსაც კი მათი აღქმა. მრავალრიცხოვან პროგრამულ პაკეტებს შორის, ჩვენი მიზნებისათვის გამოვარჩიეთ პროგრამული ენა Python. ამ არჩევანს მრავალი მიზეზი განაპირობებს. უპირველესად, ის ძალზე მოქნილია, შეუდარებელია ანალიტიკური ფუნქციების, ვიზუალიზაციის შესაძლებლობებით,  განვითარების გზებითა და საშუალებებით. გააჩნია ძალზე მრავალრიცხოვანი მიმდევართა, მომხმარებელთა  წრე, ვრცლად დოკუმენტირებულია, არსებობს უამრავი ტექსტური და პროგრამული მასალა მის შესასწავლად, პერიოდულად გამოიცემა ახალი პაკეტები, რაც აფართოებს და უფრო ყოვლისმომცველს ხდის მას.  და კიდევ, რაც არანაკლებ მნიშვნელოვანია, სხვა პროგრამული პაკეტებისაგან განსხვავებით, რომლებიც საკმაოდ ძვირი ღირს, Python - აბსოლუტურად უფასოა. ის თავისუფლად შეიძლება ჩამოვტვირთოთ საიტიდან და დავაყენოთ ჩვენს კომპიუტერებზე. Python-ის საფუძვლების გაცნობის შემდეგ განვიხილავთ მონაცემთა დამუშავებისა და ანალიზის რამოდენიმე მეთოდს. სტატისტიკური ტესტების მაგალითები Python ენაზე იქნება მოყვანილი. ამავე ენაზე იქნება წარმოდგენილი ვიზუალიზაციის პროგრამული კოდიც.</p><br> 
<p style='text-align: justify;'>იმედს ვიტოვებთ, რომ მასალის სრულად შესწავლის შემდეგ მათემატიკურ-სტატისტიკური მეთოდები უკვე ისეთ რთულ, მიუწვდომელ და საშინელ ურჩხულად აღარ წარმოგიდგებათ.  თქვენც  გამოიმუშავებთ მონაცემებთან ურთიერთობის საწყის პრაქტიკულ ჩვევებს, საკუთარ საქმიანობაშიც გამოიყენებთ შეძენილ ცოდნასა და უნარებს.   დარწმუნებული ვართ, გაგიტაცებთ მონაცემთა ანალიზის, მანქანური სწავლებისა და ხელოვნური ინტელექტის საკითხები, დაგაინტერესებთ ცოდნის შემდგომი გაღრმავება და გააგრძელებთ მოგზაურობას ამ საინტერესო სამყაროში.</p>
